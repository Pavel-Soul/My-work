## Задание 1
Дана функция $y(x) = x^2 - 3|x| + 2$

```python
def f(x):
    return x**2 - 3*abs(x) + 2
```

1)  Вывести значения функции в интервале `[a, b]`

```python
a, b = -10, 10
n = 101

X = np.linspace(a, b, n)
Y = list(map(f, X))
Y
```

---
2) Построить грфик функции на отрезке `[a, b]`

```python
plt.plot(X, Y)
plt.show()
```

## Задание 2
Постройте гистограмму оценок по группам и полу.\
Используйте несколько значений X на одной диаграмме для мужчин и женщин.

Пример данных:
> Среднее значение (мужчины) = (22, 30, 35, 35, 26)\
> Среднее значение (женщины) = (25, 32, 30, 35, 29)


```python
team = ['G1', 'G2', 'G3', 'G4', 'G5']
datasetM = [22, 30, 35, 35, 26]
datasetF = [25, 32, 30, 35, 29]

x_axis = np.arange(len(team))
width = .4
plt.bar(x_axis-width/2, datasetM, width=width, label='Man')
plt.bar(x_axis+width/2, datasetF, width=width, label='Woman')

plt.xticks(x_axis, team)

plt.legend()

plt.ylim(0, 50)
plt.show()
plt.show()
```
