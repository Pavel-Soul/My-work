```python
#Animal
class Animal:
    def __init__(self, species, age, weight):
        self.species = species
        self.age = age
        self.weight = weight
        self.color = None

    def display_info(self):
        print(f"Species: {self.species}")
        print(f"Age: {self.age} years")
        print(f"Weight: {self.weight} kg")
        if self.color:
            print(f"Color: {self.color}")
        else:
            print("Color: Not set")

    def is_adult(self):
        return self.age > 3

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, value):
        self._color = value
```

```python
# Создание экземпляра класса Animal
dog = Animal("Dog", 5, 10)

# Вывод информации о животном
dog.display_info()
#######
# Проверка, является ли животное взрослым
print(f"Is adult: {dog.is_adult()}")

# Установка цвета животного
dog.color = "Brown"

# Вывод информации о животном с установленным цветом
dog.display_info()
```

```python
#Movie
class Movie:
    def __init__(self, title, director, year, duration):
        self.title = title
        self.director = director
        self.year = year
        self.duration = duration
        self.genre = None

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Director: {self.director}")
        print(f"Year: {self.year}")
        print(f"Duration: {self.duration} minutes")
        if self.genre:
            print(f"Genre: {self.genre}")
        else:
            print("Genre: Not set")

    def is_long(self):
        return self.duration > 120

    @property
    def genre(self):
        return self._genre

    @genre.setter
    def genre(self, value):
        self._genre = value
```

```python
# Создание экземпляра класса Movie
avengers = Movie("Avengers: Endgame", "Anthony Russo, Joe Russo", 2019, 181)

# Вывод информации о фильме
avengers.display_info()
########
# Проверка, является ли фильм длинным
print(f"Is long: {avengers.is_long()}")

# Установка жанра фильма
avengers.genre = "Action, Adventure, Sci-Fi"

# Вывод информации о фильме с установленным жанром
avengers.display_info()
```

```python
#Bank
class BankAccount:
    def __init__(self, account_number, balance, owner):
        self.account_number = account_number
        self.balance = balance
        self.owner = owner
        self.currency = None

    def display_info(self):
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance}")
        print(f"Owner: {self.owner}")
        if self.currency:
            print(f"Currency: {self.currency}")
        else:
            print("Currency: Not set")

    def check_sufficient_funds(self, amount):
        return self.balance >= amount

    @property
    def currency(self):
        return self._currency

    @currency.setter
    def currency(self, value):
        self._currency = value
```

```python
# Создание экземпляра класса BankAccount
account = BankAccount("1234567890", 1000, "John Doe")

# Вывод информации о банковском счете
account.display_info()
############
# Проверка, достаточно ли денег на счете для выполнения операции
print(f"Is sufficient funds available: {account.check_sufficient_funds(500)}")

# Установка типа валюты
account.currency = "USD"

# Вывод информации о банковском счете с установленным типом валюты
account.display_info()
```

```python
#Recip
class Recipe:
    def __init__(self, name, ingredients, steps):
        self.name = name
        self.ingredients = ingredients
        self.steps = steps
        self.cuisine = None

    def display_info(self):
        print(f"Recipe Name: {self.name}")
        print("Ingredients:")
        for ingredient in self.ingredients:
            print(f"- {ingredient}")
        print("Steps:")
        for index, step in enumerate(self.steps, start=1):
            print(f"{index}. {step}")
        if self.cuisine:
            print(f"Cuisine: {self.cuisine}")
        else:
            print("Cuisine: Not set")

    def check_all_ingredients_available(self, available_ingredients):
        return all(ingredient in available_ingredients for ingredient in self.ingredients)

    @property
    def cuisine(self):
        return self._cuisine

    @cuisine.setter
    def cuisine(self, value):
        self._cuisine = value
```

```python
# Создание экземпляра класса Recipe
recipe = Recipe("Spaghetti Carbonara", ["Spaghetti", "Bacon", "Eggs", "Parmesan Cheese", "Black Pepper"], ["Cook the spaghetti according to package instructions.", "Fry the bacon until crispy.", "Beat the eggs and mix in grated Parmesan cheese.", "Drain the cooked spaghetti and mix it with the egg mixture.", "Add the crispy bacon and season with black pepper."])

# Вывод информации о рецепте
recipe.display_info()
#######
# Проверка, есть ли все ингредиенты для приготовления блюда
available_ingredients = ["Spaghetti", "Bacon", "Eggs", "Parmesan Cheese", "Black Pepper", "Milk"]
print(f"All ingredients available: {recipe.check_all_ingredients_available(available_ingredients)}")

# Установка типа кухни
recipe.cuisine = "Italian"

# Вывод информации о рецепте с установленным типом кухни
recipe.display_info()
```

# Задачи на создание иерархии классов


```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def info(self):
        print("Brand:", self.brand)
        print("Model:", self.model)
        print("Year:", self.year)


class PassengerCar(Car):
    def __init__(self, brand, model, year, engine_capacity, fuel_type):
        super().__init__(brand, model, year)
        self.engine_capacity = engine_capacity
        self.fuel_type = fuel_type

    def info(self):
        super().info()
        print("Engine Capacity:", self.engine_capacity)
        print("Fuel Type:", self.fuel_type)

    def is_suitable(self, engine_capacity_condition, fuel_type_condition):
        return self.engine_capacity == engine_capacity_condition and self.fuel_type == fuel_type_condition


class Truck(Car):
    def __init__(self, brand, model, year, payload_capacity):
        super().__init__(brand, model, year)
        self.payload_capacity = payload_capacity

    def info(self):
        super().info()
        print("Payload Capacity:", self.payload_capacity)

    def is_suitable(self, payload_capacity_condition):
        return self.payload_capacity >= payload_capacity_condition


class PassengerVehicle(Car):
    def __init__(self, brand, model, year, seat_count):
        super().__init__(brand, model, year)
        self.seat_count = seat_count

    def info(self):
        super().info()
        print("Seat Count:", self.seat_count)

    def is_suitable(self, seat_count_condition):
        return self.seat_count >= seat_count_condition


# Example usage:

cars = [
    PassengerCar("Toyota", "Camry", 2018, 2.5, "gasoline"),
    Truck("Volvo", "FH16", 2020, 20000),
    PassengerVehicle("Mercedes-Benz", "Sprinter", 2019, 20)
]

# Print full information about all cars
for car in cars:
    car.info()
    print()

# Search cars with specified characteristics
print("Searching for cars with engine capacity 2.5 and fuel type gasoline:")
for car in cars:
    if isinstance(car, PassengerCar) and car.is_suitable(2.5, "gasoline"):
        car.info()
        print()

print("Searching for trucks with payload capacity at least 15000:")
for car in cars:
    if isinstance(car, Truck) and car.is_suitable(15000):
        car.info()
        print()

print("Searching for passenger vehicles with seat count at least 15:")
for car in cars:
    if isinstance(car, PassengerVehicle) and car.is_suitable(15):
        car.info()
        print()
```

```python
class Student:
    def __init__(self, last_name, first_name, patronymic, group, course, average_grade):
        self.last_name = last_name
        self.first_name = first_name
        self.patronymic = patronymic
        self.group = group
        self.course = course
        self.average_grade = average_grade

    def info(self):
        print("Last Name:", self.last_name)
        print("First Name:", self.first_name)
        print("Patronymic:", self.patronymic)
        print("Group:", self.group)
        print("Course:", self.course)
        print("Average Grade:", self.average_grade)


class Bachelor(Student):
    def __init__(self, last_name, first_name, patronymic, group, course, average_grade):
        super().__init__(last_name, first_name, patronymic, group, course, average_grade)

    def info(self):
        super().info()


class Master(Student):
    def __init__(self, last_name, first_name, patronymic, group, course, average_grade, thesis_topic):
        super().__init__(last_name, first_name, patronymic, group, course, average_grade)
        self.thesis_topic = thesis_topic

    def info(self):
        super().info()
        print("Thesis Topic:", self.thesis_topic)

    def matches_criteria(self, average_grade_condition=None, thesis_topic_condition=None):
        if average_grade_condition and thesis_topic_condition:
            return self.average_grade >= average_grade_condition and self.thesis_topic == thesis_topic_condition
        elif average_grade_condition:
            return self.average_grade >= average_grade_condition
        elif thesis_topic_condition:
            return self.thesis_topic == thesis_topic_condition
        else:
            return True


# Пример использования:

students = [
    Bachelor("Smith", "John", "James", "Group A", 3, 4.5),
    Bachelor("Johnson", "Emily", "Anne", "Group B", 2, 3.8),
    Master("Williams", "David", "Robert", "Group C", 1, 4.2, "Data Analysis")
]

# Вывод полной информации о всех студентах
for student in students:
    student.info()
    print()

# Поиск студентов с заданными критериями
print("Поиск студентов с средним баллом не менее 4.0:")
for student in students:
    if student.matches_criteria(4.0):
        student.info()
        print()

print("Поиск магистров с темой диссертации 'Data Analysis':")
for student in students:
    if isinstance(student, Master) and student.matches_criteria(thesis_topic_condition="Data Analysis"):
        student.info()
        print()
```

```python
class Product:
    def __init__(self, name, price, manufacturer):
        self.name = name
        self.price = price
        self.manufacturer = manufacturer

    def info(self):
        print("Name:", self.name)
        print("Price:", self.price)
        print("Manufacturer:", self.manufacturer)


class Meat(Product):
    def __init__(self, name, price, manufacturer, expiration_date):
        super().__init__(name, price, manufacturer)
        self.expiration_date = expiration_date

    def info(self):
        super().info()
        print("Expiration Date:", self.expiration_date)

    def is_suitable(self, expiration_date_condition):
        return self.expiration_date >= expiration_date_condition


class Vegetables(Product):
    def __init__(self, name, price, manufacturer, seasonal):
        super().__init__(name, price, manufacturer)
        self.seasonal = seasonal

    def info(self):
        super().info()
        print("Seasonal:", self.seasonal)

    def is_suitable(self, seasonal_condition):
        return self.seasonal == seasonal_condition


class Fruits(Product):
    def __init__(self, name, price, manufacturer, seasonal):
        super().__init__(name, price, manufacturer)
        self.seasonal = seasonal

    def info(self):
        super().info()
        print("Seasonal:", self.seasonal)

    def is_suitable(self, seasonal_condition):
        return self.seasonal == seasonal_condition


# Пример использования:

products = [
    Meat("Beef", 10.99, "Farm Fresh Meats", "2023-06-30"),
    Vegetables("Carrot", 1.99, "Green Fields Farm", "Winter"),
    Fruits("Apple", 0.99, "Orchard Delights", "Fall")
]

# Вывод полной информации о всех продуктах
for product in products:
    product.info()
    print()

# Поиск продуктов с заданными условиями
print("Поиск продуктов с ценой не более 1.0:")
for product in products:
    if product.price <= 1.0:
        product.info()
        print()

print("Поиск овощей с сезонностью 'Winter':")
for product in products:
    if isinstance(product, Vegetables) and product.is_suitable("Winter"):
        product.info()
        print()
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print("Name:", self.name)
        print("Age:", self.age)


class Cat(Animal):
    def __init__(self, name, age, breed, color):
        super().__init__(name, age)
        self.breed = breed
        self.color = color

    def info(self):
        super().info()
        print("Breed:", self.breed)
        print("Color:", self.color)

    def is_suitable(self, breed_condition=None, color_condition=None):
        if breed_condition and color_condition:
            return self.breed == breed_condition and self.color == color_condition
        elif breed_condition:
            return self.breed == breed_condition
        elif color_condition:
            return self.color == color_condition
        else:
            return True


class Dog(Animal):
    def __init__(self, name, age, breed, size):
        super().__init__(name, age)
        self.breed = breed
        self.size = size

    def info(self):
        super().info()
        print("Breed:", self.breed)
        print("Size:", self.size)

    def is_suitable(self, breed_condition=None, size_condition=None):
        if breed_condition and size_condition:
            return self.breed == breed_condition and self.size == size_condition
        elif breed_condition:
            return self.breed == breed_condition
        elif size_condition:
            return self.size == size_condition
        else:
            return True


class Bird(Animal):
    def __init__(self, name, age, species, color):
        super().__init__(name, age)
        self.species = species
        self.color = color

    def info(self):
        super().info()
        print("Species:", self.species)
        print("Color:", self.color)

    def is_suitable(self, species_condition=None, color_condition=None):
        if species_condition and color_condition:
            return self.species == species_condition and self.color == color_condition
        elif species_condition:
            return self.species == species_condition
        elif color_condition:
            return self.color == color_condition
        else:
            return True


# Пример использования:

animals = [
    Cat("Fluffy", 3, "Persian", "White"),
    Dog("Buddy", 5, "Golden Retriever", "Medium"),
    Bird("Tweety", 2, "Canary", "Yellow")
]

# Вывод полной информации о всех животных
for animal in animals:
    animal.info()
    print()

# Поиск животных с заданными условиями
print("Поиск кошек с породой 'Persian':")
for animal in animals:
    if isinstance(animal, Cat) and animal.is_suitable(breed_condition="Persian"):
        animal.info()
        print()

print("Поиск птиц с окрасом 'Yellow':")
for animal in animals:
    if isinstance(animal, Bird) and animal.is_suitable(color_condition="Yellow"):
        animal.info()
        print()
```

```python
class Movie:
    def __init__(self, title, year, director, actors):
        self.title = title
        self.year = year
        self.director = director
        self.actors = actors

    def info(self):
        print("Title:", self.title)
        print("Year:", self.year)
        print("Director:", self.director)
        print("Actors:", ", ".join(self.actors))


class Comedy(Movie):
    def __init__(self, title, year, director, actors):
        super().__init__(title, year, director, actors)

    def is_suitable(self, year_condition=None, director_condition=None):
        if year_condition and director_condition:
            return self.year == year_condition and self.director == director_condition
        elif year_condition:
            return self.year == year_condition
        elif director_condition:
            return self.director == director_condition
        else:
            return True


class Drama(Movie):
    def __init__(self, title, year, director, actors):
        super().__init__(title, year, director, actors)

    def is_suitable(self, year_condition=None, director_condition=None):
        if year_condition and director_condition:
            return self.year == year_condition and self.director == director_condition
        elif year_condition:
            return self.year == year_condition
        elif director_condition:
            return self.director == director_condition
        else:
            return True


class SciFi(Movie):
    def __init__(self, title, year, director, actors):
        super().__init__(title, year, director, actors)

    def is_suitable(self, year_condition=None, director_condition=None):
        if year_condition and director_condition:
            return self.year == year_condition and self.director == director_condition
        elif year_condition:
            return self.year == year_condition
        elif director_condition:
            return self.director == director_condition
        else:
            return True


# Пример использования:

movies = [
    Comedy("Superbad", 2007, "Greg Mottola", ["Jonah Hill", "Michael Cera"]),
    Drama("The Shawshank Redemption", 1994, "Frank Darabont", ["Tim Robbins", "Morgan Freeman"]),
    SciFi("Blade Runner", 1982, "Ridley Scott", ["Harrison Ford", "Rutger Hauer"])
]

# Вывод полной информации о всех фильмах
for movie in movies:
    movie.info()
    print()

# Поиск фильмов с заданными условиями
print("Поиск комедий с годом выпуска 2007:")
for movie in movies:
    if isinstance(movie, Comedy) and movie.is_suitable(year_condition=2007):
        movie.info()
        print()

print("Поиск драм с режиссером 'Frank Darabont':")
for movie in movies:
    if isinstance(movie, Drama) and movie.is_suitable(director_condition="Frank Darabont"):
        movie.info()
        print()
```

# Задачи на создание класса и использование специальных методов

```python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color

    def area(self):
        return 3.14 * (self.radius ** 2)

    def circumference(self):
        return 2 * 3.14 * self.radius

    def __str__(self):
        return f"Круг радиуса {self.radius} и цвета {self.color}"


circle1 = Circle(5, "синий")
print(circle1.area())  # Выводит площадь круга
print(circle1.circumference())  # Выводит длину окружности круга
print(circle1)  # Выводит информацию о круге
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def __eq__(self, other):
        return self.area() == other.area()

rectangle1 = Rectangle(4, 6)
rectangle2 = Rectangle(3, 8)

print(rectangle1.area())  # Выводит площадь прямоугольника 1
print(rectangle1.perimeter())  # Выводит периметр прямоугольника 1

print(rectangle2.area())  # Выводит площадь прямоугольника 2
print(rectangle2.perimeter())  # Выводит периметр прямоугольника 2

print(rectangle1 == rectangle2)  # Сравнивает прямоугольники по площади
```

```python
class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def average_grade(self):
        if len(self.grades) > 0:
            return sum(self.grades) / len(self.grades)
        else:
            return 0

    def is_successful(self):
        return self.average_grade() > 4

    def __repr__(self):
        return f"Студент {self.name}, возраст {self.age}"

student1 = Student("Иванов", 20, [4, 5, 3, 5, 4])
student2 = Student("Петров", 19, [4, 4, 4, 4, 4])

print(student1.average_grade())  # Выводит средний балл студента 1
print(student1.is_successful())  # Проверяет успеваемость студента 1

print(student2.average_grade())  # Выводит средний балл студента 2
print(student2.is_successful())  # Проверяет успеваемость студента 2

print(student1)  # Выводит информацию о студенте 1
print(student2)  # Выводит информацию о студенте 2
```

```python
class Client:
    def __init__(self, first_name, last_name, account_number, balance):
        self.first_name = first_name
        self.last_name = last_name
        self.account_number = account_number
        self.balance = balance

    def __str__(self):
        return f"Клиент: {self.first_name} {self.last_name}, Номер счета: {self.account_number}, Баланс: {self.balance}"


class Bank:
    def __init__(self, name, address):
        self.name = name
        self.address = address
        self.clients = []

    def add_client(self, client):
        self.clients.append(client)

    def remove_client(self, client):
        if client in self.clients:
            self.clients.remove(client)

    def __str__(self):
        client_info = "\n".join(str(client) for client in self.clients)
        return f"Банк '{self.name}', адрес - {self.address}\nКлиенты:\n{client_info}"

bank = Bank("Мой Банк", "ул. Центральная, 123")

client1 = Client("Иван", "Иванов", "123456789", 5000)
client2 = Client("Петр", "Петров", "987654321", 8000)

bank.add_client(client1)
bank.add_client(client2)

print(bank)  # Выводит информацию о банке и клиентах

bank.remove_client(client1)

print(bank)  # Выводит информацию о банке и клиентах после удаления клиента
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def __str__(self):
        return f"Книга '{self.title}' автора {self.author}, издана в {self.year}"

    def __eq__(self, other):
        return self.year == other.year

book1 = Book("Война и мир", "Лев Толстой", 1869)
book2 = Book("Преступление и наказание", "Федор Достоевский", 1869)
book3 = Book("Анна Каренина", "Лев Толстой", 1877)

print(book1)  # Выводит информацию о книге 1
print(book2)  # Выводит информацию о книге 2
print(book3)  # Выводит информацию о книге 3

print(book1 == book2)  # Сравнивает книги 1 и 2 по году издания
print(book1 == book3)  # Сравнивает книги 1 и 3 по году издания
```

#  работу со стеком или очередью

```python
#СТЕК
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def peek(self):
        if not self.is_empty():
            return self.items[-1]

    def size(self):
        return len(self.items)
```

```python
#четный
def remove_non_even(stack):
    temp_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        if item % 2 == 0:
            temp_stack.push(item)
    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())
```

```python
#сумма
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def sum_primes(stack):
    total_sum = 0
    temp_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        if is_prime(item):
            total_sum += item
        temp_stack.push(item)
    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())
    return total_sum
```

```python
#квадрат
def contains_square(stack):
    temp_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        temp_stack.push(item)
        if item ** 0.5 in [x for x in temp_stack.items]:
            return True
    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())
    return False
```

```python
# делители
def remove_non_divisors(stack):
    last_item = stack.peek()
    temp_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        if last_item % item == 0:
            temp_stack.push(item)
    while not temp_stack.is_empty():
        stack.push(temp_stack.pop())
```

```python
stack = Stack()

# Добавление элементов в стек
stack.push(5)
stack.push(10)
stack.push(3)
stack.push(7)
stack.push(6)
print(stack.items)

# Удаление из стека всех элементов, которые не являются четными числами
remove_non_even(stack)
print("Стек после удаления нечетных чисел:", stack.items)  # Вывод: [10, 6]

# Нахождение суммы всех элементов стека, которые являются простыми числами
sum_of_primes = sum_primes(stack)
print("Сумма простых чисел в стеке:", sum_of_primes)  # Вывод: 0

# Проверка, содержит ли стек хотя бы один элемент, являющийся квадратом другого элемента стека
contains_square_value = contains_square(stack)
print("Стек содержит квадрат другого элемента:", contains_square_value)  # Вывод: False

# Удаление из стека всех элементов, которые не являются делителями последнего элемента стека
remove_non_divisors(stack)
print("Стек после удаления элементов-неделителей:", stack.items)  # Вывод: [10, 6]
```

```python
#ОЧЕрЕДь
#Больше заданного
class QueueGreaterThan:
    def __init__(self, limit):
        self.limit = limit
        self.queue = []

    def enqueue(self, item):
        if item > self.limit:
            self.queue.append(item)

    def dequeue(self):

        if not self.is_empty():
            return self.queue.pop(0)

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

```

```python
#уникальные элементы
class QueueUnique:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        if item not in self.queue:
            self.queue.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)
```

```python
#INT
class QueueSortedInt:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        if isinstance(item, int):
            self.queue.append(item)
            self.queue.sort(reverse=True)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)
```

```python
# Пример использования класса QueueGreaterThan
queue1 = QueueGreaterThan(5)
queue1.enqueue(10)
queue1.enqueue(3)
queue1.enqueue(7)
queue1.enqueue(2)
print(queue1.queue)  # Вывод: [10, 7]

# Пример использования класса QueueUnique
queue2 = QueueUnique()
queue2.enqueue(10)
queue2.enqueue(5)
queue2.enqueue(10)
queue2.enqueue(7)
print(queue2.queue)  # Вывод: [10, 5, 7]

# Пример использования класса QueueSortedInt
queue3 = QueueSortedInt()
queue3.enqueue(5)
queue3.enqueue(10)
queue3.enqueue(3)
queue3.enqueue(8.5)  # Не добавится, так как не является целым числом
queue3.enqueue(7)
print(queue3.queue)  # Вывод: [10, 7, 5, 3]
```

# двусвязный список

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def add(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def __str__(self):
        values = []
        current_node = self.head
        while current_node:
            values.append(str(current_node.value))
            current_node = current_node.next
        return " <-> ".join(values)

#находит максимальный элемент
def find_and_remove_max(doubly_linked_list):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return

    max_value = float("-inf")
    current_node = doubly_linked_list.head
    max_node = None

    while current_node is not None:
        if current_node.value > max_value:
            max_value = current_node.value
            max_node = current_node
        current_node = current_node.next

    if max_node is not None:
        if max_node.prev is not None:
            max_node.prev.next = max_node.next
        if max_node.next is not None:
            max_node.next.prev = max_node.prev
        if max_node == doubly_linked_list.head:
            doubly_linked_list.head = max_node.next
        if max_node == doubly_linked_list.tail:
            doubly_linked_list.tail = max_node.prev

#находит произведение квадратов
def calculate_square_product(doubly_linked_list):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return 0

    product = 1
    current_node = doubly_linked_list.head

    while current_node is not None:
        product *= current_node.value ** 2
        current_node = current_node.next

    return product

#находит количество элементов
def count_elements(doubly_linked_list):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return 0

    count = 0
    current_node = doubly_linked_list.head

    while current_node is not None:
        count += 1
        current_node = current_node.next

    return count

#палиндромом
def is_palindrome(doubly_linked_list):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return False

    forward_node = doubly_linked_list.head
    backward_node = doubly_linked_list.tail

    while forward_node is not None and backward_node is not None:
        if forward_node.value != backward_node.value:
            return False
        forward_node = forward_node.next
        backward_node = backward_node.prev

    return True

# находит минимальный элемент
def find_min(doubly_linked_list):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return None

    min_value = float("inf")
    current_node = doubly_linked_list.head

    while current_node is not None:
        if current_node.value < min_value:
            min_value = current_node.value
        current_node = current_node.next

    return min_value

#удалить элемент с указанным индексом
def remove_by_index(doubly_linked_list, index):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return

    current_node = doubly_linked_list.head
    count = 0

    while current_node is not None:
        if count == index:
            if current_node.prev is not None:
                current_node.prev.next = current_node.next
            if current_node.next is not None:
                current_node.next.prev = current_node.prev
            if current_node == doubly_linked_list.head:
                doubly_linked_list.head = current_node.next
            if current_node == doubly_linked_list.tail:
                doubly_linked_list.tail = current_node.prev
            break
        current_node = current_node.next
        count += 1

#хождения элемента в список или –1
def find_index_of_value(doubly_linked_list, value):
    if doubly_linked_list is None or doubly_linked_list.head is None:
        return -1

    current_node = doubly_linked_list.head
    index = 0

    while current_node is not None:
        if current_node.value == value:
            return index
        current_node = current_node.next
        index += 1

    return -1
```

```python
# Пример использования функции find_and_remove_max
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(5)
doubly_linked_list.add(10)
doubly_linked_list.add(3)
doubly_linked_list.add(7)

print("Список до удаления максимального элемента:", doubly_linked_list)

find_and_remove_max(doubly_linked_list)

print("Список после удаления максимального элемента:", doubly_linked_list)


# Пример использования функции calculate_square_product
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(2)
doubly_linked_list.add(3)
doubly_linked_list.add(4)

product = calculate_square_product(doubly_linked_list)

print("Произведение квадратов всех элементов:", product)


# Пример использования функции count_elements
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(5)
doubly_linked_list.add(10)
doubly_linked_list.add(3)
doubly_linked_list.add(7)

count = count_elements(doubly_linked_list)

print("Количество элементов в списке:", count)


# Пример использования функции is_palindrome
doubly_linked_list1 = DoublyLinkedList()
doubly_linked_list1.add(2)
doubly_linked_list1.add(4)
doubly_linked_list1.add(6)
doubly_linked_list1.add(4)
doubly_linked_list1.add(2)

print("Список 1:", doubly_linked_list1)
print("Является ли список 1 палиндромом:", is_palindrome(doubly_linked_list1))

doubly_linked_list2 = DoublyLinkedList()
doubly_linked_list2.add(1)
doubly_linked_list2.add(2)
doubly_linked_list2.add(3)
doubly_linked_list2.add(4)

print("Список 2:", doubly_linked_list2)
print("Является ли список 2 палиндромом:", is_palindrome(doubly_linked_list2))


# Пример использования функции find_min
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(5)
doubly_linked_list.add(10)
doubly_linked_list.add(3)
doubly_linked_list.add(7)

min_value = find_min(doubly_linked_list)

print("Минимальный элемент в списке:", min_value)


# Пример использования функции remove_by_index
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(5)
doubly_linked_list.add(10)
doubly_linked_list.add(3)
doubly_linked_list.add(7)

print("Список до удаления элемента по индексу:", doubly_linked_list)

remove_by_index(doubly_linked_list, 1)

print("Список после удаления элемента по индексу:", doubly_linked_list)


# Пример использования функции find_index_of_value
doubly_linked_list = DoublyLinkedList()
doubly_linked_list.add(5)
doubly_linked_list.add(10)
doubly_linked_list.add(3)
doubly_linked_list.add(7)

index = find_index_of_value(doubly_linked_list, 3)

print("Индекс элемента 3 в списке:", index)
```

# сортирует

```python
#сортирует его по алфавиту
def insertion_sort_words(word_list):
    for i in range(1, len(word_list)):
        key = word_list[i]
        j = i - 1
        while j >= 0 and word_list[j] > key:
            word_list[j + 1] = word_list[j]
            j -= 1
        word_list[j + 1] = key
    return word_list

words = ["banana", "apple", "cherry", "orange"]
sorted_words = insertion_sort_words(words)
print(sorted_words)
```

```python
#класс «Товар»
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __repr__(self):
        return f"{self.name}: {self.price}"

    @staticmethod
    def quicksort_products(products):
        def partition(arr, low, high):
            i = low - 1
            pivot = arr[high].price
            for j in range(low, high):
                if arr[j].price <= pivot:
                    i += 1
                    arr[i], arr[j] = arr[j], arr[i]
            arr[i + 1], arr[high] = arr[high], arr[i + 1]
            return i + 1

        def quicksort_helper(arr, low, high):
            if low < high:
                pi = partition(arr, low, high)
                quicksort_helper(arr, low, pi - 1)
                quicksort_helper(arr, pi + 1, high)

        quicksort_helper(products, 0, len(products) - 1)

product_list = [Product("Apple", 2.5), Product("Banana", 1.2), Product("Orange", 3.0), Product("Cherry", 1.8)]
print("Список товаров до сортировки:")
for product in product_list:
    print(product)

Product.quicksort_products(product_list)

print("Список товаров после сортировки по цене:")
for product in product_list:
    print(product)
```

```python
#список дат
def bubble_sort_dates(date_list):
    n = len(date_list)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if date_list[j] > date_list[j + 1]:
                date_list[j], date_list[j + 1] = date_list[j + 1], date_list[j]
    return date_list

dates = ["2022-01-15", "2022-02-10", "2022-01-05", "2022-03-01"]
sorted_dates = bubble_sort_dates(dates)
print(sorted_dates)
```

```python
#список строк
def selection_sort_strings(string_list):
    n = len(string_list)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if len(string_list[j]) < len(string_list[min_index]):
                min_index = j
        string_list[i], string_list[min_index] = string_list[min_index], string_list[i]
    return string_list

strings = ["apple", "banana", "grape", "orange", "cherry"]
sorted_strings = selection_sort_strings(strings)
print(sorted_strings)
```

```python
#«Клиент»
class Client:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"{self.name}: {self.age}"

    @staticmethod
    def bubble_sort_clients(client_list):
        n = len(client_list)
        for i in range(n - 1):
            for j in range(n - 1 - i):
                if client_list[j].age > client_list[j + 1].age:
                    client_list[j], client_list[j + 1] = client_list[j + 1], client_list[j]

clients = [Client("John", 25), Client("Alice", 32), Client("Bob", 20)]
print("Список клиентов до сортировки:")
for client in clients:
    print(client)

Client.bubble_sort_clients(clients)

print("Список клиентов после сортировки по возрасту:")
for client in clients:
    print(client)
```

```python
#список чисел
def insertion_sort_descending(number_list):
    for i in range(1, len(number_list)):
        key = number_list[i]
        j = i - 1
        while j >= 0 and number_list[j] < key:
            number_list[j + 1] = number_list[j]
            j -= 1
        number_list[j + 1] = key
    return number_list

numbers = [5, 2, 9, 1, 7]
sorted_numbers = insertion_sort_descending(numbers)
print(sorted_numbers)
```

```python
#«Заказ»
class Заказ:
    def __init__(self, order_id, date):
        self.order_id = order_id
        self.date = date

    def __repr__(self):
        return f"Order ID: {self.order_id}, Date: {self.date}"

    @staticmethod
    def quick_sort_orders(order_list):
        def partition(orders, low, high):
            pivot = orders[high].date
            i = low - 1
            for j in range(low, high):
                if orders[j].date <= pivot:
                    i += 1
                    orders[i], orders[j] = orders[j], orders[i]
            orders[i + 1], orders[high] = orders[high], orders[i + 1]
            return i + 1

        def quick_sort_helper(orders, low, high):
            if low < high:
                pi = partition(orders, low, high)
                quick_sort_helper(orders, low, pi - 1)
                quick_sort_helper(orders, pi + 1, high)

        quick_sort_helper(order_list, 0, len(order_list) - 1)

orders = [Заказ(1, "2022-02-10"), Заказ(2, "2022-01-15"), Заказ(3, "2022-03-01")]
print("Список заказов до сортировки:")
for order in orders:
    print(order)

Заказ.quick_sort_orders(orders)

print("Список заказов после сортировки по дате:")
for order in orders:
    print(order)
```

#  деревом

```python
#Найти наибольший элемент, меньший заданного значения, в бинарном дереве поиска.
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def find_largest_smaller_value(root, target):
    result = None
    while root:
        if root.value < target:
            result = root.value
            root = root.right
        else:
            root = root.left
    return result

# Создание бинарного дерева поиска
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.right.right = TreeNode(14)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right.left = TreeNode(13)

target_value = 9
largest_smaller_value = find_largest_smaller_value(root, target_value)
print("Наибольший элемент, меньший", target_value, ":", largest_smaller_value)
```

```python
#Найти высоту бинарного дерева поиска.
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def find_tree_height(root):
    if root is None:
        return 0
    left_height = find_tree_height(root.left)
    right_height = find_tree_height(root.right)
    return max(left_height, right_height) + 1

# Создание бинарного дерева поиска
root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(9)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.right.left = TreeNode(7)
root.right.right = TreeNode(10)
root.right.right.right = TreeNode(11)

tree_height = find_tree_height(root)
print("Высота бинарного дерева:", tree_height)
```

```python
# книга
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __repr__(self):
        return f"Book: {self.title} by {self.author}"


class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash_function(self, author):
        return hash(author) % self.size

    def add_book(self, book):
        index = self._hash_function(book.author)
        self.table[index].append(book)

    def find_books_by_author(self, author):
        index = self._hash_function(author)
        return self.table[index]

    def __str__(self):
        table_str = ""
        for i, cell in enumerate(self.table):
            table_str += f"Cell {i}: "
            if len(cell) > 0:
                table_str += ", ".join([book.title for book in cell])
            else:
                table_str += "Empty"
            table_str += "\n"
        return table_str

book1 = Book("Book 1", "Author A")
book2 = Book("Book 2", "Author B")
book3 = Book("Book 3", "Author A")
book4 = Book("Book 4", "Author C")

hash_table = HashTable(3)
hash_table.add_book(book1)
hash_table.add_book(book2)
hash_table.add_book(book3)
hash_table.add_book(book4)

print(hash_table)

author = "Author A"
books_by_author = hash_table.find_books_by_author(author)
print(f"Books by {author}:")
for book in books_by_author:
    print(book)
```
