```python
import numpy as np
from itertools import permutations
import matplotlib.pyplot as plt
import random as rand
```

```python
def graph(func, test_count=1_000_000):
    rand.seed(1223)
    results = [0]

    for _ in range(test_count):
        results.append(func() + results[-1])

    results = [el / (n+1) for n, el in enumerate(results[1:])]

    mean = sum(results)/test_count
    s = results[test_count//50:]
    k = max(s) - min(s)
    plt.ylim(mean-k, mean+k)

    real_mean = sum(results[test_count//20:]) / (test_count//20*19)

    plt.plot(results)
    plt.plot([0, test_count], [real_mean, real_mean])
    return real_mean
```

```python
teams = ['a'] + 2*['p'] + 7*['t']
```

```python
def teamA():
    te = rand.sample(teams, 5)
    return te.count('p') == 1

graph(teamA)
```

```python
def teamB():
    te = rand.sample(teams, 5)
    return te.count('a') == 1 and te.count('p') == 0
graph(teamB)
```

```python
def teamC():
    te = rand.sample(teams, 5)
    return te.count('p') == 2 and te.count('a') == 1

graph(teamC)
```

```python
def homeA():
  floors = list(np.random.randint(2, 8, 6))
  return not (2 in floors) and not (3 in floors) and not (4 in floors)


graph(homeA)
```

```python
def homeB():
  floors = list(np.random.randint(2, 8, 6))
  return floors.count(7) == 3


graph(homeB)
```

```python
def homeC():
  floors = list(np.random.randint(2, 8, 6))
  return len(set(floors)) == 6


graph(homeC)
```

```python
def homeD():
  return len(set(rand.randint(1, 6) for _ in range(6))) == 1


graph(homeD)
```

```python
from math import factorial


def samp(a,b):
    count = min(a, b)
    while True:
        if a % count == b % count == 0:
            a //= count
            b //= count
        else:
            count -= 1
        if count <= 1:
            return a, b

def C(n, k):
    # return '{}/{}'.format(*samp(factorial(n), factorial(k)*factorial(n-k)))
    return factorial(n) / (factorial(k)*factorial(n-k))
```
