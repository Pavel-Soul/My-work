```python
# 6.3 а
X = rv_discrete(values=(np.arange(12), np.ones(12)/12))
Y = rv_discrete(values=(np.arange(13), np.ones(13)/13))
Z = rv_discrete(values=([3, 7], [0.9, 0.1]))

s = 0
for x in range(12):
    for y in range(13):
        for z in (3, 7):
            if x + y + z == 12:
                s += X.pmf(x) * Y.pmf(y) * Z.pmf(z)
s
```
