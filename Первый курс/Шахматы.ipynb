```python
class Rook:  #Ладья
  def __init__(self, color, moved = False, attacked = False): #Параметры Ладьи
    self.color = color
    self.moved = moved
    self.attacked = attacked

  def Available_moves(self, n_row, n_col, Board): #Нахождение доступных ходов для ладьи
    moves = []

    d_x = 1  #Нахождение дост ходов по горизонтали направо
    while (n_row + d_x) < 9:
      if type(Board.board[n_row + d_x - 1][n_col - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col))
      elif (Board.board[n_row + d_x - 1][n_col - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col))
        break
      d_x += 1

    d_x = -1 #Нахождение доступных ходов по горизонтали налево
    while (n_row + d_x) > 0:
      if type(Board.board[n_row + d_x - 1][n_col -1]) == Empty_square:
        moves.append((n_row + d_x, n_col))
      elif (Board.board[n_row + d_x - 1][n_col - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col))
        break
      d_x -= 1

    d_y = 1 #Нахождение доступных ходов по вертикали вверх
    while (n_col + d_y) < 9:
      if type(Board.board[n_row - 1][n_col + d_y - 1]) == Empty_square:
        moves.append((n_row, n_col + d_y))
      elif (Board.board[n_row - 1][n_col + d_y - 1]).color == self.color:
        break
      else:
        moves.append((n_row, n_col + d_y))
        break
      d_y += 1

    d_y = -1 #Нахождение доступных ходов по горизонтали вверх
    while (n_col + d_y) > 0:
      if type(Board.board[n_row - 1][n_col + d_y - 1]) == Empty_square:
        moves.append((n_row, n_col + d_y))
      elif (Board.board[n_row - 1][n_col + d_y - 1]).color == self.color:
        break
      else:
        moves.append((n_row, n_col + d_y))
        break
      d_y -= 1
    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    return self.Available_moves(n_row, n_col, Board)
```

```python
class Knight: #Конь
  def __init__(self, color, moved = False, attacked = False): #Параметры коня
    self.color = color
    self.moved = moved
    self.attacked = attacked

  def Available_moves(self, n_row, n_col, Board): #Нахождение доступных ходов для коня
    moves = []
    for d_x in (-2, -1, 1, 2):
      for d_y in (-2, -1, 1, 2):
        if abs(d_x) != abs(d_y) and (0 < (n_row + d_x) < 9) and (0 < (n_col + d_y) < 9):
          if  type(Board.board[n_row + d_x - 1][n_col + d_y - 1]) == Empty_square:
            moves.append((n_row+d_x, n_col+d_y))
          elif (Board.board[n_row + d_x - 1][n_col + d_y - 1]).color != self.color:
            moves.append((n_row+d_x, n_col+d_y))
    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    return self.Available_moves(n_row, n_col, Board)
```

```python
class Bishop:
  def __init__(self, color, moved = False, attacked = False):
    self.color = color
    self.moved = moved
    self.attacked = attacked

  def Available_moves(self, n_row, n_col, Board):
    moves = []

    d_x = 1
    while  (n_row + d_x) < 9 and (n_col + d_x) < 9:
      if type(Board.board[n_row + d_x - 1][n_col + d_x - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col + d_x))
      elif (Board.board[n_row + d_x - 1][n_col + d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col + d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row + d_x) < 9 and (n_col - d_x) > 0:
      if type(Board.board[n_row + d_x - 1][n_col - d_x - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col - d_x))
      elif (Board.board[n_row + d_x - 1][n_col - d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col - d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row - d_x) > 0 and (n_col - d_x) > 0:
      if type(Board.board[n_row - d_x - 1][n_col - d_x - 1]) == Empty_square:
        moves.append((n_row - d_x, n_col - d_x))
      elif (Board.board[n_row - d_x - 1][n_col - d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row - d_x, n_col - d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row - d_x) > 0 and (n_col + d_x) < 9:
      if type(Board.board[n_row - d_x - 1][n_col + d_x - 1]) == Empty_square:
        moves.append((n_row - d_x, n_col + d_x))
      elif (Board.board[n_row - d_x - 1][n_col + d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row - d_x, n_col + d_x))
        break
      d_x += 1

    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    return self.Available_moves(n_row, n_col, Board)
```

```python
class Queen:
  def __init__(self, color, moved = False, attacked = False):
    self.color = color
    self.moved = moved
    self.attacked = attacked
  def Available_moves(self, n_row, n_col, Board):
    moves = []

    d_x = 1
    while  (n_row + d_x) < 9 and (n_col + d_x) < 9:
      if type(Board.board[n_row + d_x - 1][n_col + d_x - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col + d_x))
      elif (Board.board[n_row + d_x - 1][n_col + d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col + d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row + d_x) < 9 and (n_col - d_x) > 0:
      if type(Board.board[n_row + d_x - 1][n_col - d_x - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col - d_x))
      elif (Board.board[n_row + d_x - 1][n_col - d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col - d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row - d_x) > 0 and (n_col - d_x) > 0:
      if type(Board.board[n_row - d_x - 1][n_col - d_x - 1]) == Empty_square:
        moves.append((n_row - d_x, n_col - d_x))
      elif (Board.board[n_row - d_x - 1][n_col - d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row - d_x, n_col - d_x))
        break
      d_x += 1

    d_x = 1
    while  (n_row - d_x) > 0 and (n_col + d_x) < 9:
      if type(Board.board[n_row - d_x - 1][n_col + d_x - 1]) == Empty_square:
        moves.append((n_row - d_x, n_col + d_x))
      elif (Board.board[n_row - d_x - 1][n_col + d_x - 1]).color == self.color:
        break
      else:
        moves.append((n_row - d_x, n_col + d_x))
        break
      d_x += 1

    d_x = 1
    while (n_row + d_x) < 9:
      if type(Board.board[n_row + d_x - 1][n_col - 1]) == Empty_square:
        moves.append((n_row + d_x, n_col))
      elif (Board.board[n_row + d_x - 1][n_col - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col))
        break
      d_x += 1

    d_x = -1
    while (n_row + d_x) > 0:
      if type(Board.board[n_row + d_x - 1][n_col -1]) == Empty_square:
        moves.append((n_row + d_x, n_col))
      elif (Board.board[n_row + d_x - 1][n_col - 1]).color == self.color:
        break
      else:
        moves.append((n_row + d_x, n_col))
        break
      d_x -= 1

    d_y = 1
    while (n_col + d_y) < 9:
      if type(Board.board[n_row - 1][n_col + d_y - 1]) == Empty_square:
        moves.append((n_row, n_col + d_y))
      elif (Board.board[n_row - 1][n_col + d_y - 1]).color == self.color:
        break
      else:
        moves.append((n_row, n_col + d_y))
        break
      d_y += 1

    d_y = -1
    while (n_col + d_y) > 0:
      if type(Board.board[n_row - 1][n_col + d_y - 1]) == Empty_square:
        moves.append((n_row, n_col + d_y))
      elif (Board.board[n_row - 1][n_col + d_y - 1]).color == self.color:
        break
      else:
        moves.append((n_row, n_col + d_y))
        break
      d_y -= 1

    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    return self.Available_moves(n_row, n_col, Board)
```

```python
class Pawn:
  def __init__(self, color, moved = False, attacked = False,
               R_enpassant = False, L_enpassant = False):
    self.color = color
    self.moved = moved
    self.attacked = attacked
    self.R_enpassant = R_enpassant
    self.L_enpassant = L_enpassant

  def Available_moves(self, n_row, n_col, Board):
    moves = []

    if self.color == 'w':
      if type(Board.board[n_row - 1 - 1][n_col - 1]) == Empty_square:
        moves.append((n_row - 1, n_col))
        if not self.moved and type(Board.board[n_row - 2 - 1][n_col - 1]) == Empty_square:
          moves.append((n_row - 2, n_col))
      if n_col + 1 < 9:
        if self.R_enpassant or (Board.board[n_row - 1 - 1][n_col + 1 - 1].color == 'b'):
          moves.append((n_row-1, n_col + 1))
      if n_col - 1 > 0:
        if self.L_enpassant or (Board.board[n_row - 1 - 1][n_col - 1 - 1].color == 'b'):
          moves.append((n_row-1, n_col - 1))
      # if ((self.R_enpassant and n_col != 8)
      #     or Board.board[n_row - 1 - 1][n_col + 1 - 1].color == 'b'):
      #   moves.append((n_row - 1, n_col + 1))
      # if ((self.L_enpassant and n_col != 1)
      #     or Board.board[n_row - 1 - 1][n_col - 1 - 1].color == 'b'):
      #   moves.append((n_row - 1, n_col - 1))

    else:
      if type(Board.board[n_row + 1 - 1][n_col - 1]) == Empty_square:
        moves.append((n_row+1, n_col))
        if not self.moved and type(Board.board[n_row + 2 -1][n_col - 1]) == Empty_square:
          moves.append((n_row+2, n_col))
      if n_col + 1 < 9:
        if self.R_enpassant or (Board.board[n_row + 1 - 1][n_col + 1 - 1].color == 'b'):
          moves.append((n_row+1, n_col + 1))
      if n_col - 1 > 0:
        if self.L_enpassant or (Board.board[n_row + 1 - 1][n_col - 1 - 1].color == 'b'):
          moves.append((n_row+1, n_col - 1))
      # if ((self.R_enpassant and n_col != 8)
      #     or Board.board[n_row + 1 - 1][n_col + 1 - 1].color == 'w'):
      #   moves.append((n_row + 1, n_col + 1))
      # if ((self.L_enpassant and n_col != 1)
      #    or Board.board[n_row + 1 - 1][n_col - 1 - 1].color == 'w'):
      #   moves.append((n_row + 1, n_col - 1))
    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    a_squares = []
    if self.color == 'b':
      if n_row + 1 < 9 and n_col + 1 < 9:
        a_squares.append((n_row + 1, n_col + 1))
      if n_row + 1 < 9 and n_col - 1 > 0:
        a_squares.append((n_row+1, n_col - 1))
    else:
      if n_row - 1 < 9 and n_col + 1 < 9:
        a_squares.append((n_row - 1, n_col + 1))
      if n_row - 1 < 9 and n_col - 1 > 0:
        a_squares.append((n_row - 1, n_col -1))
    return a_squares
```

```python
class Empty_square:
  def __init__(self, color = '', attacked = False):
    self.color = color
    self.moved = False
    self.attacked = attacked
  def Available_moves(self, n_row, n_col, Board):
    return []
  def Attacked_squares(self, n_row, n_col, Board):
    return []
```

```python
class King:
  def __init__(self, color, moved = False, attacked = False):
    self.color = color
    self.moved = moved
    self.attacked = attacked

  def Available_moves(self, n_row, n_col, Board):
    moves = []

    for d_x in range(-1, 2):
      for d_y in range(-1, 2):
        if (0 < n_row + d_x < 9) and (0 < n_col + d_y < 9):
          if Board.board[n_row + d_x - 1][n_col + d_y - 1].attacked == False:
            if Board.board[n_row + d_x - 1][n_col + d_y - 1].color != self.color:
              moves.append((n_row + d_x, n_col + d_y))
    return moves

  def Attacked_squares(self, n_row, n_col, Board):
    a_squares = []
    for d_x in range(-1, 2):
      for d_y in range(-1, 2):
        if ((d_x != 0 or d_y != 0) and
            (0 < n_row + d_x < 9) and
            (0 < n_col + d_y < 9)and
            type(Board.board[d_x-1][d_y-1]) == Empty_square):
          a_squares.append((n_row + d_x, n_col + d_y))
    return a_squares
```

```python
class Chessboard:
  def __init__(self,  board = ''):
    self.board = board

  def clear(self):
    self.board = [[Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()],
                  [Empty_square(), Empty_square(), Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(), Empty_square(), Empty_square()]]

    def set_board(self):
        board = [[Rook('b'), Knight('b'), Bishop('b'), Queen('b'),
                  King('b'), Bishop('b'),  Knight('b'), Rook('b')],
                  [Pawn('b'), Pawn('b'), Pawn('b'), Pawn('b'),
                  Pawn('b'), Pawn('b'), Pawn('b'), Pawn('b')],
                  [Empty_square(), Empty_square(),Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(),Empty_square(), Empty_square(),],
                  [Empty_square(), Empty_square(),Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(),Empty_square(), Empty_square(),],
                  [Empty_square(), Empty_square(),Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(),Empty_square(), Empty_square(),],
                  [Empty_square(), Empty_square(),Empty_square(), Empty_square(),
                  Empty_square(), Empty_square(),Empty_square(), Empty_square(),],
                  [Pawn('w'), Pawn('w'), Pawn('w'), Pawn('w'),
                  Pawn('w'), Pawn('w'), Pawn('w'), Pawn('w')],
                  [Rook('w'), Knight('w'), Bishop('w'),  Queen('w'),
                  King('w'), Bishop('w'),  Knight('w'), Rook('w')]]
        self.board = board

  def available_moves_of_piece(self, n_row, n_col):
    Piece = self.board[n_row-1][n_col-1]
    moves = Piece.Available_moves(n_row, n_col, self)
    available_moves = []
    for move in moves:
      if self.check_the_move(n_row, n_col, move[0], move[1], Piece.color):
        available_moves.append(move)
    return available_moves

  def insert_Piece(self, Piece, n_row, n_col):
    self.board[n_row-1][n_col-1] = Piece

  def delete_Piece(self, n_row, n_col):
    self.board[n_row-1][n_col-1] = Empty_square()

  def coords_of_king(self, color):
    for row_index, row in enumerate(self.board):
      for el_index, el in enumerate(row):
        if type(el) == King and el.color == color:
          return (row_index+1, el_index+1)

  def check_the_move(self, n_row_start, n_col_start, n_row_end, n_col_end, color):
    holded_piece = self.board[n_row_end-1][n_col_end-1]
    self.insert_Piece(self.board[n_row_start-1][n_col_start-1], n_row_end, n_col_end)
    self.delete_Piece(n_row_start, n_col_start)
    if color == 'w':
      self.refresh_attacked_squares_pieces('b')
    else:
      self.refresh_attacked_squares_pieces('w')
    row_King, col_King = self.find_coords_of_King(color)
    if self.board[row_King-1][col_King-1].attacked == True:
      self.insert_Piece(self.board[n_row_end-1][n_col_end-1], n_row_start, n_col_start)
      self.delete_Piece(n_row_end, n_col_end)
      self.insert_Piece(holded_piece, n_row_end, n_col_end)
      return False
    else:
      self.insert_Piece(self.board[n_row_end-1][n_col_end-1], n_row_start, n_col_start)
      self.delete_Piece(n_row_end, n_col_end)
      self.insert_Piece(holded_piece, n_row_end, n_col_end)
      return True


  def move_Piece(self, start_row, start_col, end_row, end_col, color):
    self.refresh_attacked_squares_pieces(self.board[end_row-1][end_col-1].color)
    if (end_row, end_col) in self.available_moves_of_piece(start_row, start_col):
      some_piece = self.board[start_row-1][start_col-1]
      some_piece.moved = True
      self.refresh_attacked_squares_pieces(color)
      self.insert_Piece(some_piece, end_row, end_col)
      self.delete_Piece(start_row, start_col)
      self.promoting_pawns(color)
      self.refresh_attacked_squares_pieces(color)
      self.refresh_enpassant(start_row, start_col, end_row, end_col, color)
      return 0
    else:
      return 1

  def find_coords_of_King(self, color):
    for row in self.board:
      for piece in row:
        if type(piece) == King and piece.color == color:
          return self.board.index(row)+1, row.index(piece) + 1


  def castling(self, inputt, color):
    if color == 'w' and inputt == '0-0':
      if (type(self.board[8-1][5-1]) == King and
          self.board[8-1][5-1].color == 'w' and
          not self.board[8-1][5-1].moved and
          type(self.board[8-1][6-1]) == Empty_square and
          self.board[8-1][6-1].attacked == False and
          type(self.board[8-1][7-1]) == Empty_square and
          self.board[8-1][7-1].attacked == False and
          type(self.board[8-1][8-1]) == Rook and
          self.board[8-1][5-1].color == 'w' and
          not self.board[8-1][5-1].moved):

            self.insert_Piece(self.board[8-1][5-1], 8, 7)
            self.board[8-1][7-1].moved = True
            self.delete_Piece(8,5)
            self.insert_Piece(self.board[8-1][8-1], 8, 6)
            self.board[8-1][6-1].moved = True
            self.delete_Piece(8,8)
            self.refresh_attacked_squares_pieces('w')
            return 0
      else:
            return 1
    elif color == 'w' and inputt == '0-0-0':
      if (type(self.board[8-1][5-1]) == King and
          self.board[8-1][5-1].color == 'w' and
          not self.board[8-1][5-1].moved and
          type(self.board[8-1][4-1]) == Empty_square and
          self.board[8-1][4-1].attacked == False and
          type(self.board[8-1][3-1]) == Empty_square and
          self.board[8-1][3-1].attacked == False and
          type(self.board[8-1][1-1]) == Rook and
          self.board[8-1][1-1].color == 'w' and
          not self.board[8-1][1-1].moved):
            self.insert_Piece(self.board[8-1][5-1], 8, 3)
            self.board[8-1][3-1].moved = True
            self.delete_Piece(8,5)
            self.insert_Piece(self.board[8-1][8-1], 8, 4)
            self.board[8-1][4-1].moved = True
            self.delete_Piece(8,1)
            self.refresh_attacked_squares_pieces('w')
            return 0
      else:
            return 1
    elif color == 'b' and inputt == '0-0':
      if (type(self.board[1-1][5-1]) == King and
          self.board[1-1][5-1].color == 'b' and
          not self.board[1-1][5-1].moved and
          type(self.board[1-1][6-1]) == Empty_square and
          self.board[1-1][6-1].attacked == False and
          type(self.board[1-1][7-1]) == Empty_square and
          self.board[1-1][7-1].attacked == False and
          type(self.board[1-1][8-1]) == Rook and
          self.board[1-1][5-1].color == 'b' and
          not self.board[1-1][5-1].moved):

            self.insert_Piece(self.board[1-1][5-1], 1, 7)
            self.board[1-1][7-1].moved = True
            self.delete_Piece(1,5)
            self.insert_Piece(self.board[1-1][8-1], 1, 6)
            self.board[1-1][6-1].moved = True
            self.delete_Piece(1,8)
            self.refresh_attacked_squares_pieces('b')
            return 0
      else:
            return 1
    else:
      if (type(self.board[1-1][5-1]) == King and
          self.board[1-1][5-1].color == 'b' and
          not self.board[1-1][5-1].moved and
          type(self.board[1-1][4-1]) == Empty_square and
          self.board[1-1][4-1].attacked == False and
          type(self.board[1-1][3-1]) == Empty_square and
          self.board[1-1][3-1].attacked == False and
          type(self.board[1-1][1-1]) == Rook and
          self.board[1-1][1-1].color == 'b' and
          not self.board[1-1][1-1].moved):
            self.insert_Piece(self.board[1-1][5-1], 1, 3)
            self.board[1-1][3-1].moved = True
            self.delete_Piece(1,5)
            self.insert_Piece(self.board[1-1][8-1], 1, 4)
            self.board[1-1][4-1].moved = True
            self.delete_Piece(1,1)
            self.refresh_attacked_squares_pieces('b')
            return 0
      else:
            return 1

  def promoting_pawns(self, color):
    if color == 'w':
      for index_el, el in enumerate(self.board[0]):
        if type(el) == Pawn:
          print("You've promoted pawn! Choose a Piece:")
          inputt = input()
          Pieces = {'Rook': Rook('w', True,), 'Queen': Queen('w'),
                    'Knight': Knight('w'), 'Bishop': Bishop('w')}
          self.insert_Piece(Pieces[inputt], 1, index_el + 1)
    else:
      for index_el, el in enumerate(self.board[7]):
        if type(el) == Pawn:
          print("You've promoted pawn! Choose a Piece:")
          inputt = input()
          Pieces = {'Rook': Rook('b', True,), 'Queen': Queen('b'),
                    'Knight': Knight('b'), 'Bishop': Bishop('b')}
          self.insert_Piece(Pieces[inputt], 7, index_el + 1)

  def input_to_coords(self, inputt, color):
      if inputt[0].lower() != inputt[0]:
        inputt = inputt[1:]
      if '-' in inputt:
        inputt = inputt.split('-')
      elif 'x' in inputt:
        inputt = inputt.split('x')
      dic = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h':8}
      start = list(inputt[0])
      end = list(inputt[1])
      start_col = dic[start[0]]
      start_row = 9-int(start[1])
      end_col = dic[end[0]]
      end_row = 9 - int(end[1])
      if self.board[start_row-1][start_col-1].color == color:
        return start_row, start_col, end_row, end_col
      else:
        return -1

  def refresh_attacked_squares_pieces(self, color):
    attacked_squares = set()
    for row_index, row in enumerate(self.board):
      for el_index, el in enumerate(row):
        if el.color == color:
          for coords in el.Attacked_squares(row_index+1, el_index+1, self):
            attacked_squares = attacked_squares | set([coords])
    attacked_squares = list(attacked_squares)
    for n_row in range(1, 9):
      for n_col in range(1, 9):
        if (n_row, n_col) in attacked_squares:
          self.board[n_row-1][n_col-1].attacked = True
        else:
          self.board[n_row-1][n_col-1].attacked = False

  def refresh_enpassant(self, start_row, start_col, end_row, end_col, color):
    if type(self.board[end_row-1][end_col-1]) == Pawn:
      if (end_col - 1 > 0) and end_row - start_row == 2:
          if (type(self.board[end_row-1][end_col-1 - 1]) == Pawn and
          self.board[end_row - 1][end_col-1 - 1].color != color):
            self.board[end_row-1][end_col-1 - 1].R_enpassant = True
      elif (end_col + 1 < 9) and end_row - start_row == 2:
          if (type(self.board[end_row-1][end_col+1 - 1]) == Pawn and
          self.board[end_row - 1][end_col+1 - 1].color != color):
            self.board[end_row-1][end_col+1 - 1].L_enpassant = True

  def refresh_enpassant_at_the_end_of_turn(self, color):
    for row in self.board:
      for el in row:
        if type(el) == Pawn and el.color == color:
          el.L_enpassant = False
          el.R_enpassant = False


  def check_the_check(self, color):
    king = self.find_coords_of_King(color)
    return self.board[king[0]-1][king[1]-1].attacked

  def check_able_to_move(self, color):
    for x in range(1, 9):
      for y in range(1, 9):
        if color == 'w':
          self.refresh_attacked_squares_pieces('b')
        if color == 'b':
          self.refresh_attacked_squares_pieces('w')
        if (len(self.available_moves_of_piece(x, y)) != 0 and
            self.board[x-1][y-1].color == color):
          return True
    return False
  pieces = {(King, 'w'): '♚', (King, 'b'): '♔',
              (Queen, 'w'): '♛', (Queen, 'b'): '♕',
              (Rook, 'w'): '♜', (Rook, 'b'): '♖',
              (Bishop, 'w'): '♝', (Bishop, 'b'): '♗',
              (Knight, 'w'): '♞', (Knight, 'b'): '♘',
              (Pawn, 'w'): '♟', (Pawn, 'b'): '♙',
              (Empty_square, ''): '. '}
  def show_attacked_pieces(self, color):
    new_board = []
    cnt = 9
    self.refresh_attacked_squares_pieces(color)
    for row in self.board:
      cnt -= 1
      new_board.append([str(cnt)])
      for el in row:
        if el.attacked == True and type(el) != Empty_square:
          new_board[-1].append('× ')
        elif el.attacked == True and type(el) == Empty_square:
          new_board[-1].append('o ')
        else:
          new_board[-1].append(self.pieces[(type(el), el.color)])
    new_board.append(['  a ', 'b ', 'c ', 'd ', 'e ', 'f ', 'g ', 'h '])
    for row in new_board:
      print(' '.join(row))

  def show(self):
    new_board = []
    cnt = 9
    for row in self.board:
      cnt -= 1
      new_board.append([str(cnt)])
      for el in row:
        new_board[-1].append(self.pieces[(type(el), el.color)])
    new_board.append(['  a ', 'b ', 'c ', 'd ', 'e ', 'f ', 'g ', 'h '])
    for row in new_board:
      print(' '.join(row))
```

```python
def search_unmoved_Rooks(Board, color):
  Rooks = []
  for index_row, row in enumerate(Board.board):
    for index_el, el in enumerate(row):
      if type(el) == Rook and el.color == color and el.moved == False:
        Rooks.append((index_row+1, index_el+1))
  return Rooks
```

```python
def start_chess(inputts = []):
  cnt = 1
  Board = Chessboard()
  Board.set_board()
  Board.show()
  Board.refresh_attacked_squares_pieces('w')
  turns = []
  inputs = []
  turns.append([[i for i in el] for el in Board.board])
  while True:
    if cnt % 2 == 1:
      color = 'w'

      if not Board.check_able_to_move(color):
        king_coords = Board.find_coords_of_King(color)
        king = Board.board[king_coords[0]-1][king_coords[1]-1]
        if king.attacked == True:
          print('Black won!')
        else:
          print('Tie')
        break
      if not inputts:
        print("White's Turn")

    else:
      color = 'b'

      if not Board.check_able_to_move(color):
        king_coords = Board.find_coords_of_King(color)
        king = Board.board[king_coords[0]-1][king_coords[1]-1]
        if king.attacked == True:
          print('White won!')
        else:
          print('Tie')
        break
      if not inputts:
        print("Black's Turn")


    if Board.check_the_check(color):
      print('Check!')

    if not inputts:
      inputt = input()
    else:
      if cnt == len(inputts)+1:
        break
      if input() == '>':
        inputt = inputts[cnt-1]

    if inputt == 'attacked pieces' or inputt == 'ap':
      Board.show_attacked_pieces(color)
      continue
    if inputt == 'break':
      saved_board = Board.board
      return turns
    if inputt == '<':
      if len(turns) > 1:
        Board.board = [[i for i in el] for el in turns[-2]]
        color = 'w' if color == 'b' else 'b'
        start_row, start_col, end_row, end_col = Board.input_to_coords(inputs[-1], color)
        Board.board[start_row-1][start_col-1].moved = False
        turns = turns[:-1]
        inputs = inputs[:-1]
        cnt -= 1
        Board.show()
      else:
        print("It's first turn")
      continue

      Board.board[start_row-1][start_col-1].moved = True
    if inputt not in ('0-0', '0-0-0'):
      if Board.input_to_coords(inputt, color) == -1:
        print("It's not your Piece")
        continue
      else:
        start_row, start_col, end_row, end_col = Board.input_to_coords(inputt, color)
      if Board.move_Piece(start_row, start_col, end_row, end_col, color):
        print("It's illegal move")
        continue
    else:
      if Board.castling(inputt, color):
        print("It's illegal move")
        continue
    inputs.append(inputt)
    turns.append([[i for i in el] for el in Board.board])
    Board.refresh_enpassant_at_the_end_of_turn(color)
    Board.show()
    cnt += 1
```
