```python
axes = [5, 5, 5]
# Создаем объемный массив 5 на 5 на 5 и там где должен быть элемент ставим True
data = np.ones(axes, dtype=bool)

ax = plt.axes(projection='3d')

# Отображаем
ax.voxels(data)
plt.show()
```

```python
import matplotlib.patches as mpatches
polygon = mpatches.RegularPolygon((0.5, 0.5), numVertices=8, radius=0.4, color='g')
ax = plt.axes()
ax.set_aspect(aspect=1)
ax.add_patch(polygon)
plt.show()
```

```python
fig = plt.figure(facecolor='Black')
ax = plt.axes(projection='3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)


def sphere(r, x0=0, y0=0, z0=0, col=None):
    x = r * np.outer(np.cos(u), np.sin(v)) + x0
    y = r * np.outer(np.sin(u), np.sin(v)) + y0
    z = r * np.outer(np.ones(np.size(u)), np.cos(v)) + z0
    ax.plot_surface(x, y, z,  rstride=2, cstride=2, color=col)


r = 5
for x in range(4):
    for y in range(4):
        for z in range(4):
            if (x + y + z) % 2 == 0:
                sphere(3, x*r*1.5, y*r*1.5, z*r*1.5, col='#724390')
            else:
                sphere(5, x*r*1.5, y*r*1.5, z*r*1.5, col='#2bb64e')

plt.show()
```
