# Упражнение 2. Анимация в pygame.

```python
# Инициализация Pygame
pygame.init()

# Установка размеров окна
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Движение фигур")

# Определение цветов
black = (0, 0, 0)
red = (255, 0, 0)
blue = (0, 0, 255)
green = (0, 255, 0)
yellow = (255, 255, 0)

# Определение параметров для фигур
square_side = 50
rectangle_width = 80
rectangle_height = 40
circle_radius = 30
triangle_height = 50
triangle_base = 50

# Установка начальных координат для фигур
square_x, square_y = 50, 200
rectangle_x, rectangle_y = 200, 200
circle_x, circle_y = 350, 200
triangle_x, triangle_y = 500, 200

# Установка скорости движения
speed = 5

# Основной игровой цикл
running = True
while running:
    screen.fill('white')

    # Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Обновление координат для движущихся фигур
    square_x += speed
    rectangle_x += speed
    circle_x += speed
    triangle_x += speed

    # Отрисовка фигур
    pygame.draw.rect(screen, red, (square_x, square_y, square_side, square_side))
    pygame.draw.rect(screen, blue, (rectangle_x, rectangle_y, rectangle_width, rectangle_height))
    pygame.draw.circle(screen, green, (circle_x, circle_y), circle_radius)
    pygame.draw.polygon(screen, yellow, [(triangle_x, triangle_y), (triangle_x + triangle_base, triangle_y),
                                       (triangle_x + triangle_base / 2, triangle_y - triangle_height)])

    # Проверка столкновения с границами экрана и изменение направления движения
    if (circle_x + circle_radius > width or square_x < 0 or triangle_x + triangle_base > width or rectangle_x + rectangle_width > width):
        speed = -speed

    # Обновление экрана
    pygame.display.flip()
    pygame.time.Clock().tick(30)

# Завершение работы Pygame
pygame.quit()
sys.exit()
```
