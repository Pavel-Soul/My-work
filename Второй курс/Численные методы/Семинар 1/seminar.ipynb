```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def solve2(f, accuracy=0.01, steps=10_000, start=0, line=10):
    k = start
    for step in range(steps):
        # Проверяем найдено ли решение
        if abs(f(k)) <= accuracy:
            return k
        if abs(f(k + line)) < abs(f(k - line)):
            k += line
        else:
            k -= line
        line /= 2
    return k
```

```python
def draw(f, t=10, ans=None):
    X = np.linspace(-t, t + 1, 100)
    Y = f(X)
    plt.axvline(0, -t, t + 1, c="red")
    plt.axhline(0, -t, t + 1, c="red")
    plt.plot(X, Y)
    if ans is not None:
        plt.scatter(ans, f(ans), c='orange', marker='o', s=100)
```

```python
def solve(f, accuracy=0.01, a=-1000, b=1000, steps = 10_000):
    while steps:
        steps -= 1
        k = (a+b)/2
        t = f(k)
        if abs(t) <= accuracy:
            return k
        if t > 0:
            b = k
        else:
            a = k
    return k
```

```python
def draw_and_solve(f, a=-1000, b=1000):
    t = solve(f, a=a, b=b)
    draw(f, t * 5, t)
    print(f"Root: {t}")
```

```python
# 1

draw_and_solve(lambda x: x**3 + 2 * x + 2)
```

```python
# 2
draw_and_solve(lambda x: x**3 + 15 * x - 10)
```

```python
# 3
draw_and_solve(lambda x: np.log(x) - np.sin(x))
```

```python
# 4
draw_and_solve(lambda x: np.cos(x) - x, a=5, b=-5)
```

```python
# 5
draw_and_solve(lambda x: np.sin(x) - (1-x**2)**0.5, a=0, b=1)
```
