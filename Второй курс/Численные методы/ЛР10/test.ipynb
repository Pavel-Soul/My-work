```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Генерация примерного сигнала: синусоидальная волна с шумом
fs = 1000  # Частота дискретизации
t = np.linspace(0, 1, fs, endpoint=False)
signal = (
    np.sin(2 * np.pi * 50 * t)
    + np.sin(2 * np.pi * 120 * t)
    + 0.5 * np.random.randn(t.size)
)

# Применение быстрого преобразования Фурье (БПФ)
signal_fft = fft(signal)
frequencies = fftfreq(t.size, 1 / fs)

# Построение спектра частот
plt.plot(frequencies[: t.size // 2], np.abs(signal_fft)[: t.size // 2])
plt.title("Спектр частот сигнала")
plt.xlabel("Частота (Гц)")
plt.ylabel("Амплитуда")
plt.show()
```
