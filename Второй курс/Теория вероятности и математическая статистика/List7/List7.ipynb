```python
from math import e, gamma, pi
from sympy.stats import variance
import numpy as np
from scipy.optimize import minimize
```

```python
import sympy as sp

# Определение символов
theta = sp.symbols("theta", positive=True)

# Заданные вероятности
p_minus_1 = 2 * theta
p_1 = 2 * theta
p_2 = theta
p_0 = 1 - 5 * theta

# Лог-правдоподобие
log_likelihood = sp.prod([p_minus_1, p_1, p_2, p_0])

# Находим производную лог-правдоподобия по theta
derivative = sp.diff(log_likelihood, theta)

# Решаем уравнение для нахождения оценки максимального правдоподобия
mle_theta = sp.solve(derivative, theta)[0]

# Вывод результата
print("Оценка параметра theta по методу максимального правдоподобия:", mle_theta)
```

```python
# Вычисление математического ожидания оценки
expectation_mle = sp.integrate(mle_theta, (theta, 0, 1))

# Вывод результата
print("Математическое ожидание оценки:", expectation_mle)
# Оценка состоятельна
```

```python
# Оценка дисперсии оценки
variance_mle = variance(mle_theta)
n = sp.symbols("n", positive=True, integer=True)
# Вывод результата
print("Дисперсия оценки:", variance_mle)

# Предел дисперсии по вероятности при n -> ∞
variance_limit = sp.limit(variance_mle, n, sp.oo)

# Вывод результата
print("Предел дисперсии по вероятности при n -> ∞:", variance_limit)
# Оценка не смещена, тк дисперсия остатков сходится к 0
```

```python
def f(a, b, x):
    return b / pi / (b**2 + (x - a) ** 2)



# Определите лог-правдоподобие для распределения Коши
def log_likelihood(params, data):
    a, b = params
    return -np.sum(np.log(f(a, b, data)))


# Начальные значения параметров
initial_params = [0, 1]

# Минимизация отрицательного лог-правдоподобия
result = minimize(log_likelihood, initial_params, args=(data,))

# Выведите оценки параметров
print("Оценка параметра a:", result.x[0])
print("Оценка параметра b:", result.x[1])
```
