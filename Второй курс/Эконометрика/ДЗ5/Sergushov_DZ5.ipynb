```python
#  данные X (темп прироста зарплаты) и Y (темп прироста безработицы )
X = [530.16, 545.98, 552.03, 526.32, 594.75, 557.53, 643.63, 524.82, 447.75, 385.54, 487.23, 632.77]
Y = [7.2, 7.2, 7.1, 7.3, 6.5, 8.5, 8.4, 9.2, 9.8, 11.6, 8.8, 7]
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan, het_white, het_goldfeldquandt
from scipy.stats import bartlett

# данные X (темп прироста зарплаты) и Y (темп прироста безработицы )
X = [530.16, 545.98, 552.03, 526.32, 594.75, 557.53, 643.63, 524.82, 447.75, 385.54, 487.23, 632.77]
Y = [7.2, 7.2, 7.1, 7.3, 6.5, 8.5, 8.4, 9.2, 9.8, 11.6, 8.8, 7]

# 1. Получите остатки модели регрессии Y на X
X = sm.add_constant(X)  # Добавляем константу (intercept) к X
model = sm.OLS(Y, X).fit()  # Строим регрессию Y на X
residuals = model.resid  # Получаем остатки
print("Остатки регрессии:")
print(residuals)

# 2. Проведем 3 теста на гомоскедастичность

# Тест Бройша-Пагана
bp_test_stat, bp_p_value, _, _ = het_breuschpagan(residuals, X)
print(f"Тест Бройша-Пагана: статистика = {bp_test_stat}, p-value = {bp_p_value}")

# Тест Уайта
white_test_stat, white_p_value, _, _ = het_white(residuals, X)
print(f"Тест Уайта: статистика = {white_test_stat}, p-value = {white_p_value}")

# Тест Гольдфельда-Квандта
gq_test_stat, gq_p_value, _ = het_goldfeldquandt(residuals, X)
print(f"Тест Гольдфельда-Квандта: статистика = {gq_test_stat}, p-value = {gq_p_value}")

# Тест Барлетта
bartlett_test_stat, bartlett_p_value = bartlett(Y, residuals)
print(f"Тест Барлетта: статистика = {bartlett_test_stat}, p-value = {bartlett_p_value}")

```

##№2
Тест Бройша-Пагана (Breusch-Pagan Test):

Условия применения: Тест Бройша-Пагана применяется для проверки гомоскедастичности в множественной регрессии. Нулевая гипотеза состоит в том, что остатки регрессии являются гомоскедастичными, то есть дисперсия ошибок постоянна. Если p-значение теста меньше выбранного уровня значимости, то нулевая гипотеза отвергается в пользу альтернативной гипотезы о наличии гетероскедастичности.

Источник: Григорьева, Т. Л. (2013). Эконометрика: Учебное пособие. Москва: Дело.

Тест Уайта (White Test):

Условия применения: Тест Уайта также используется для проверки гомоскедастичности в множественной регрессии. Нулевая гипотеза заключается в том, что остатки регрессии являются гомоскедастичными. Если p-значение теста меньше уровня значимости, то нулевая гипотеза отвергается.

Источник: Демешев, Б. Б., & Малиновская, О. О. (2007). Эконометрика: Учебное пособие. Москва: Издательство "Экономика".

Тест Гольдфельда-Квандта (Goldfeld-Quandt Test):

Условия применения: Тест Гольдфельда-Квандта используется для проверки гомоскедастичности в регрессиях с изменяющейся дисперсией остатков в зависимости от значений независимых переменных. Нулевая гипотеза предполагает гомоскедастичность остатков. Если p-значение меньше уровня значимости, то нулевая гипотеза отвергается.

Источник: Гурьев, С. В. (2016). Эконометрика. Москва: Издательство "Эксмо".

Тест Барлетта (Bartlett Test):

Условия применения: Тест Барлетта обычно используется в анализе дисперсии (ANOVA) для проверки гомоскедастичности. Нулевая гипотеза заключается в том, что дисперсии в различных группах данных равны. Если p-значение теста меньше уровня значимости, то нулевая гипотеза отвергается.

Источник: Ширяев, А. Н. (1980). Вероятность. Москва: Наука
